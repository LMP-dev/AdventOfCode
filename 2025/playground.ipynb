```python
s="[.##.] (3) (1,3) (2) (2,3) (0,2) (0,1) {3,5,4,7}"
l, *b, j = s.split()
```

```python
buttons_schematic = []
for str_schematic in b:
    print(f"Working on scheme: {str_schematic}")
    scheme = eval(str_schematic)
    if type(scheme) is int:
        scheme = tuple([scheme])
    button_scheme = np.zeros(len(l)-2, dtype=bool)
    for index in scheme:
        print(f"working on scheme {str_schematic} at the button position {index}")
        button_scheme[index] = True
    buttons_schematic.append(button_scheme)

print(f"The buttons schematics are: {buttons_schematic}")
```
